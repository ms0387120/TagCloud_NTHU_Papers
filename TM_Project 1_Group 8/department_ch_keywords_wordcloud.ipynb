# Department_ch_keywords_wordcloud

# Step1: read excel file

```python
import xlrd
theses = xlrd.open_workbook('20170330.xls')
records= theses.sheets()[0]
```

```python
nrows= records.nrows
print (nrows)
```

# Step2: 查詢論文數最多的前五個系所外加服科所

```python
from collections import Counter
wordcount= Counter()
for i in range(1,nrows):
    wordcount[records.cell(i,2).value]+=1
wordcount.most_common(5)
```

## Step3: 系所_ch_keywords、系所_weight

```python
from nltk import word_tokenize #CS_ch_keywords (list)
CS_word=""
for i in range(1,nrows):
    if records.cell(i,2).value=="資訊工程學系":
        CS_word+=records.cell(i,17).value
CS_word_str= CS_word.replace("'","").replace("\\n"," ").replace("[","").replace("]","").replace(","," ").replace("("," ").replace(")"," ")
CS_ch_keywords= word_tokenize(CS_word_str)
print(CS_ch_keywords)
```

```python
from collections import Counter #CS_weight
CS_wordcount= Counter()
for i in CS_ch_keywords:
    CS_wordcount[i]+=1
for k,v in CS_wordcount.most_common():
    print(v, k)
```
